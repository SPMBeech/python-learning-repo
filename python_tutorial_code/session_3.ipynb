```python
# Exercise 1
# import pandas as pd
filepath = r"https://raw.githubusercontent.com/data-to-insight/ERN-sessions/main/data/ChildIdentifiers.csv"
child_data = pd.read_csv(filepath)

# child_data.info()
# child_data.head(10)
# child_data.tail(10)
```

```python
# house_prices

# datetime/datetime object
house_prices["Period_dt"] = pd.to_datetime(house_prices["Period"], format="%Y-%m") # need to use correct dat format i.e. - UK vs US
house_prices["Period_dt"].info
print
```

```python
# Exercise 2
child_data["PersonBirthDate_dt"] = pd.to_datetime(child_data["PersonBirthDate"], format="%Y-%m-%d")
child_data["ExpectedPersonBirthDate_dt"] = pd.to_datetime(child_data["ExpectedPersonBirthDate"], format="%Y-%m-%d")
child_data["PersonDeathDate_dt"] = pd.to_datetime(child_data["PersonDeathDate"], format="%Y-%m-%d", errors="coerce")
# child_data["PersonBirthDate_dt"].info
# child_data.info
child_data.info()

```

```python
# max, min, median, mean, sum etc.
max_change = house_prices["Percentage change (monthly) All property types"].max()
min_change = house_prices["Percentage change (monthly) All property types"].min()
sum_change = house_prices["Percentage change (monthly) All property types"].sum()
print(min_change)
print(max_change)
print(f"Max change: {max_change}, Min change: {min_change}")

median_price = house_prices["Average price All property types"].median()
mean_price = house_prices["Average price All property types"].mean()
print(median_price)
print(mean_price)

# slicing is making selection of rows based on defined conditions
condition = house_prices["Average price All property types"] < mean_price
less_than_mean = house_prices[condition]
less_than_mean
```

```python
#current_date = pd.to_datetime("today")
collection_end = pd.to_datetime("31/03/2025", format="%d/%m/%Y")
child_data["time_delta"] = collection_end - child_data["PersonBirthDate_dt"]
child_data["Age"] = child_data["time_delta"] / pd.Timedelta(365.25,"d") # not technically correct way to do this
child_data["Age"] = child_data["Age"].astype(int) # rounds down to closest integer
child_data
```

```python
# Exercise 3
min_age = child_data["Age"].min()
max_age = child_data["Age"].max()
mean_age = child_data["Age"].mean()
median_age = child_data["Age"].median()
mode_age = child_data["Age"].mode()

print(mode_age)
print(max_age - min_age)
```
