```python
def fxy(x=90, y=10): # set default values
    print(f'The sum of {x} and {y} is {x + y}')

fxy(x=7, y=13)
fxy(y=22, x=12) # order doesn't matter if variables stated

fxy(9, 12) # will pass values to order of variables
fxy() # will revert to default values
fxy(y=9)
```

```python
# Exercise 1

def xdivy (x, y):
    if x > y:
        print(x / y)
    else:
        print(y / x)

xdivy(10,11)
xdivy(100,20)
    
```

```python
# Getting stuff out of function
def divide_nums_return(x, y): # these variables only have their values inside the function (local scope)
    if x > y:
        z = x / y
    else:
        z = y / x
    return z

ans = divide_nums_return(10,100)
print(ans)
```

```python
# Getting multiple things passed from a function

def sum_prod(i, j):
    sum = i + j
    prod = i * j

    return sum, prod

answer = sum_prod(44, 97) # returns tuple
print(answer)
print(type(answer))

ans_sum, ans_prod = sum_prod(10, 100) # returns as multiple variables
print(ans_sum)
print(ans_prod)
print(type(ans_sum))
```

```python
# Exercise 2

def div_minus(a, b):
    if a > b:
        divide = a / b
        minus = a - b
    else:
        divide = b / a
        minus = b - a
    return divide, minus

ans_div, ans_minus = div_minus(1000,100)
print(ans_div)
print(ans_minus)


```

```python
# Another method

def div_minus(a, b):
    if a > b:
        return a / b, a -b
    else:
        return b / a, b -a 
    
ans_div, ans_minus = div_minus(1000,100)
print(ans_div)
print(ans_minus)
```

```python
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/data-to-insight/ERN-sessions/main/data/ChildCharacteristics.csv')
 
def ethnic_check(ethnicity):
    if ethnicity[0] == "W":
        return 'White'
    else:
        return 'Not white'

# . apply method passes function on a row wise basis

df['ethnic_subgroup'] = df['Ethnicity'].apply(ethnic_check)

df[['Ethnicity','ethnic_subgroup']]
```

```python
# 'https://raw.githubusercontent.com/data-to-insight/ERN-sessions/main/data/ChildIdentifiers.csv'
# write a function that calculates ages based on the data, and returns a new column in age range buckets
# 0-5, 6-11, 12-17, 18+

# import data
# write functions
# using conditionals
# chaqnge to date
# calculation on df columns
# making new column

df = pd.read_csv('https://raw.githubusercontent.com/data-to-insight/ERN-sessions/main/data/ChildIdentifiers.csv')

def age_buckets(row):
    bday = pd.to_datetime(row, format="%Y-%m-%d")
    today = pd.to_datetime('today')

    # age = (today - bday)/pd.Timedelta(365.25, 'D')

    if bday + pd.DateOffset(years=6) > today:
        return '0-5 years'
    elif bday + pd.DateOffset(years=12) > today:
        return '6-11 years'
    elif bday + pd.DateOffset(years=18) > today:
        return '12-17 years'
    else:
        return '18+'
        
df['Age Bucket'] = df['PersonBirthDate'].apply(age_buckets)

df[['PersonBirthDate', 'Age Bucket']]

```
