```python
# Loops, then loops with conditionals

import time # this is so we can use the sleep function

numbers = list(range(1,11))

for number in numbers:
    print(number)
    time.sleep(1)
```

```python
for i in numbers: # i is the iterable in the iterator, can be called whatever is most applicable
    outcome = i ** 2 ## ** squares number
    print(outcome)
    time.sleep(1)
    print("inside the loop")

print("outside the loop")
```

```python
# Conditionals - if, elif and else (similar to VB)

names_list = ['Will','Claire','Marcel','Alistair','Joe']

for n in names_list:
    if n == 'Marcel':
        print('Woof!')
    elif n == 'Will':
        print(f"Hi, I'm Will") # Need double quotes when there are single quotes in text
    else:
        print(f'Hello, my name is {n}.')
    
```

```python
# Iterating over dictionaries
dict_1 = {'key_1':1,
          'key_2':2,
          'key_3':3} # touples

dict_1.items()

for k, v in dict_1.items():
    print(f'The key is: {k}')
    print(f'The value is: {v}')

for k in dict_1.keys():
    print(dict_1[k]) # accessing the key from the dictionary
```

```python
name_job = {'Will':'Data manager',
            'Annie':'Retired',
            'Andy':'Retired',
            'Naiomi':'Policy Writer',
            'Scout':'Moral Support'}

for name, job in name_job.items():
    if name != 'Scout':
        print(f"Hi, I'm {name}, and I am a {job}")
    else:
        print('Woof!')
```

```python
# Loops in loops
evens = list(range(2,13,2))
odds = list(range(1,12,2))

multiple_list = []

for even in evens:
    for odd in odds:
        result = even * odd
        stmt = f'{even} times {odd} = {result}'
        multiple_list.append(stmt)

print(multiple_list)
        
```

```python
# % can be used as modulus - divides left number by right and returns remainder

#print(81 % 9 == 0)

# fizzbang

numbers = list(range(1,101))

for number in numbers:
    if number % 3 == 0 and number % 5 == 0:
        print('fizzbang')
    if number % 3 == 0:
        print('fizz')
    elif number % 5 == 0:
        print('bang')
    else:
        print(number)
```
