```python
# Functions and Methods
# from a user perspective they are very similar
# pre-written bits of code
# eg print(argument) is a function
# methods are associated with specific types
# eg variable_of_type.method()

# 'append(), .extend(), .remove() <- methods for lists

cols_1 = ["red", "blue","red"]

print(cols_1)

cols_1.append("brown") # mutates cols_1
print(cols_1)

cols_2 = ["purple","orange"]
cols_1.append(cols_2) # adds stuff as a single element
print(cols_1)

cols_1.extend(cols_2) # adds iterables as individual elements
print(cols_1)

cols_1.extend("brown") # adds as individual elements
print(cols_1)

cols_1.remove("purple") # removes only 1 element with a given value (not when nested though)
print(cols_1)
```
